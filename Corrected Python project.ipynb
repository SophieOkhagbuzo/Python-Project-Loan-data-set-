```python
df.columns=df.columns.str.lower()
```

```python
df.isnull().sum()
```

```python
missing_data=df.isnull().sum()
```

```python
#### Delete columns with more than 50% missing values
threshold = len(df) * 0.5
columns_to_drop = missing_data[missing_data > threshold].index
```

```python
columns_to_drop
```

```python
#Delete the columns above
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
#Fill missing values for numerical columns with the median
numerical_columns = df_cleaned.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    df_cleaned.fillna({col:df_cleaned[col].median()}, inplace=True)
```

```python
# Fill missing values for categorical columns with the mode
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_cleaned.fillna({col:df_cleaned[col].mode()}, inplace=True)
```

```python
# Convert date columns to datetime format
for col in df_cleaned.columns:
    if 'Date' in col or 'date' in col:
        df_cleaned[col] = pd.to_datetime(df_cleaned[col], format="%d/%m/%Y", errors='coerce')
```

### What is/are the main feature(s) of interest in your dataset?

>The main features of interest in my loan dataset are as follows;

1. Loan_Status: This is the primary feature of interest, especially when assessing loan performance. We need to know  whether loans are approved, rejected, paid off, or defaulted.
2. Loan_Amount: The principal amount borrowed is crucial for understanding the scale of the loan, its risk, and its impact on the lender's portfolio.
3. Interest_Rate: This determines the cost of borrowing for the borrower and the potential revenue for the lender. It is key in assessing profitability and risk.
4. Borrower Credit_Score: A critical predictor of loan repayment behavior. Higher credit scores generally indicate lower risk of default.
5. Loan Term: The duration of the loan affects the monthly payment amount and the overall interest paid over time. It’s important for understanding repayment schedules and loan affordability.
6. Monthly_Payment: This indicates the borrower's monthly financial obligation and can be used to assess affordability and the likelihood of default.
Remaining_Balance: Helps in tracking the loan’s lifecycle and understanding how much of the loan is still outstanding.
7. Borrower Income: Used to assess the borrower’s ability to repay the loan. Higher income typically suggests a lower risk of default.

## Univariate Exploration

### 1. Loan status

Distribution of loan status to understand the proportion of approved, rejected, and defaulted loans.

```python
plt.figure()
sns.countplot(data=df_cleaned, y='loanstatus', order=df_cleaned['loanstatus'].value_counts().index, hue ='loanstatus', palette ='crest')
plt.title('Loan Status')
plt.xlabel(' ')
plt.show()
```

### 2. Loan Amount
Kdeplot to visualize the range and distribution of loan amounts.

```python
plt.figure()
sns.kdeplot(df_cleaned['loanoriginalamount'], color ='teal',fill= True)
plt.title('Loan amount');
```

### 3. Credit score range( upper and lower)
Combined kde plot with hue parameter to analyze the creditworthiness of borrowers.

```python
plt.figure()
sns.kdeplot(df_cleaned['creditscorerangeupper'],color ='teal',fill= True)
sns.kdeplot(df_cleaned['creditscorerangelower'],color ='orange')
plt.title('Credit score range (upper and lower) ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

### 4.Borrower's annual percentage rate %
Distribution to see the spread of interest rates across loans.

```python
plt.figure()
sns.histplot(df_cleaned['borrowerapr'], bins=15, color ='teal')
plt.title('Borrowers annual percentage rate')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

### 5.Employment Status
Frequency distribution to assess the employment background of borrowers.

```python
plt.figure()
sns.countplot(data=df_cleaned, y='employmentstatus', order=df_cleaned['employmentstatus'].value_counts().index, hue ='employmentstatus', palette ='crest')
plt.title('Employment Status')
plt.show()
```

### Income Range
Used to assess the borrower’s ability to repay the loan. Higher income typically suggests a lower risk of default.

### Bi-Variate charts

### 1. Loan Status vs. Employment Status

shows the correlation between employment status and loan approval.

Explore how different loan terms impact loan performance.

### 4. Loan Status vs. Income

### 5. Loan Original Amount vs. Borrower APR
Understand how loan amounts influence the interest rate offered.

```python
plt.figure(figsize=(15, 5))
sns.displot(data=df_cleaned, x= 'loanoriginalamount', y='borrowerapr' )
plt.title('Loan amount vs Borrower Annual Percentage Rate')
plt.xlabel(' ')
plt.ylabel('')
plt.show()
```
